```python
import baostock as bs
import pandas as pd

#### 登陆系统 ####
lg = bs.login()
# 显示登陆返回信息
print('login respond error_code:'+lg.error_code)
print('login respond  error_msg:'+lg.error_msg)
```

```python
# 获取中证500成分股
rs = bs.query_zz500_stocks()
print('query_zz500 error_code:'+rs.error_code)
print('query_zz500  error_msg:'+rs.error_msg)
rs
```

```python
# 打印结果集
zz500_stocks = []
while (rs.error_code == '0') & rs.next():
    # 获取一条记录，将记录合并在一起
    zz500_stocks.append(rs.get_row_data())
zz500_result = pd.DataFrame(zz500_stocks, columns=rs.fields)
zz500_result
```

```python
# 获取行业分类数据
hysj = bs.query_stock_industry()
print('query_stock_industry error_code:'+hysj.error_code)
print('query_stock_industry respond  error_msg:'+hysj.error_msg)

# 打印结果集
industry_list = []
while (hysj.error_code == '0') & hysj.next():
    # 获取一条记录，将记录合并在一起
    industry_list.append(hysj.get_row_data())
classified_result = pd.DataFrame(industry_list, columns=hysj.fields)
classified_result
```

```python
inner_result = pd.merge(zz500_result, classified_result,on=['code', 'code_name'])
inner_result
```

```python
# 获取证券基本资料
rs = bs.query_stock_basic()
# rs = bs.query_stock_basic(code_name="浦发银行")  # 支持模糊查询
print('query_stock_basic respond error_code:'+rs.error_code)
print('query_stock_basic respond  error_msg:'+rs.error_msg)

# 打印结果集
data_list = []
while (rs.error_code == '0') & rs.next():
    # 获取一条记录，将记录合并在一起
    data_list.append(rs.get_row_data())
query_stock_basic_result = pd.DataFrame(data_list, columns=rs.fields)
query_stock_basic_result
```

```python
inner_result = pd.merge(inner_result, query_stock_basic_result,on=['code', 'code_name'])
inner_result
```

```python
inter_result = inner_result[inner_result['ipoDate'] <= "2013-12-31"]
inter_result
```

```python
for x in range(2012,2020):
    for i in range(1,5): 
        dupont_list = []
        for item in inter_result['code']:
            rs_dupont= bs.query_dupont_data(code=item, year=x, quarter=i)
            while (rs_dupont.error_code == '0') & rs_dupont.next():
                dupont_list.append(rs_dupont.get_row_data())
        result_profit = pd.DataFrame(dupont_list, columns=rs_dupont.fields)
        result_profit['dupontROE'] = pd.to_numeric(result_profit['dupontROE'], errors='coerce').fillna(0)#转换数据
        result_profit_temp = result_profit[['code','dupontROE']]
        qua = "_%d_%d"%(x,i)
        if(i==1 and x==2012): 
            dupont_info_dict = result_profit
        else:
            dupont_info_dict = pd.merge(dupont_info_dict, result_profit_temp,on=['code'],suffixes=("_%d_%d"%(x,i-1),qua))
        #dupont_info_dict.rename(columns={'dupontROE':qua})

        print(dupont_info_dict)
```

```python
dupont_info_dict.info()
```

```python
dupont_info_dict['2017Sum'] = dupont_info_dict.filter(regex=("dupontROE_2017_*")).sum(axis=1)
dupont_info_dict['2017Sum']
```

```python
for x in range(2012,2020): # sum of ROE of each year from 2012 - 2019
    name = "%d"%(x)
    dupont_info_dict[name+"Sum"] = dupont_info_dict.filter(regex=("dupontROE_"+name+"_*")).sum(axis=1)
dupont_info_dict
```

```python
select_compamy = dupont_info_dict
for x in range(2012,2020): # select ROE >15% for 7 years
    name = "%d"%(x)
    select_compamy = select_compamy[select_compamy[name+"Sum"]>=0.15]
```

```python
select_compamy
```

```python
get_industry_info = pd.merge(select_compamy, inter_result,on=['code'])
get_industry_info['industry'].unique() #get industry content
get_industry_info
```

```python
col_list = ['医药生物', '非银金融', '食品饮料', '休闲服务', '公用事业',
       '农林牧渔', '传媒', '电子', '商业贸易', '计算机',
       '轻工制造', '家用电器', '通信', '纺织服装']
get_industry_finish = get_industry_info[get_industry_info['industry'].isin(col_list)]#
get_industry_finish
```

```python
get_industry_finish.to_csv("D:\\stock_selection.csv", encoding="gbk", index=False)
print(get_industry_finish)
```

```python
query_history_k_data_list = []
query_history_k_data = pd.DataFrame(columns=['code', 'peTTM_20', 'peTTM_avg', 'peTTM_80', 'pbMRQ_20', 'pbMRQ_avg', 'pbMRQ_80'])
for code in get_industry_finish['code']:
    result_list = []
    rs = bs.query_history_k_data_plus(code,
        "date,code,close,peTTM,pbMRQ,psTTM,pcfNcfTTM",
        start_date='2010-08-06', end_date='2020-08-06', 
        frequency="d", adjustflag="3")
    while (rs.error_code == '0') & rs.next():
        result_list.append(rs.get_row_data())
    result = pd.DataFrame(result_list, columns=rs.fields)
    result['peTTM'] = pd.to_numeric(result['peTTM'], errors='coerce').fillna(0)
    result['pbMRQ'] = pd.to_numeric(result['pbMRQ'], errors='coerce').fillna(0)
    peTTMAvg = result['peTTM'].mean()
    pe_20 = result['peTTM'].quantile(0.20)
    pe_80 = result['peTTM'].quantile(0.80)
    pbMRQ_20 = result['pbMRQ'].quantile(0.20)
    pbMRQ_avg = result['pbMRQ'].mean()
    pbMRQ_80 = result['pbMRQ'].quantile(0.80)
    query_history_k_data_list.append([code, pe_20, peTTMAvg, pe_80, pbMRQ_20, pbMRQ_avg, pbMRQ_80])
query_history_k_data = pd.DataFrame(query_history_k_data_list, columns=['code', 'peTTM_20', 'peTTM_avg', 'peTTM_80', 'pbMRQ_20', 'pbMRQ_avg', 'pbMRQ_80'])
```
